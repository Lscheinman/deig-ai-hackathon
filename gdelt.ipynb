# Creating a HANA table

```python
import os
from dotenv import load_dotenv
load_dotenv()
from hdbcli import dbapi

# HANA Credentials
address = os.getenv("HANA_HOST")
port = os.getenv("HANA_PORT")
user = os.getenv("HANA_USER")
password = os.getenv("HANA_PASSWORD")
```

```python
# Create HANA tables for daily GDELT summary
connection = dbapi.connect(address=address, port=port, user=user, password=password)
cursor = connection.cursor()

# Drop tables if they exist
for table in ["GDELT_UA_DAILY_CAMEOS", "GDELT_UA_DAILY_SUMMARY"]:
    try:
        cursor.execute(f"DROP TABLE {table}")
        print(f"✓ Dropped {table}")
    except Exception as e:
        print(f"Table {table} doesn't exist: {str(e)[:50]}")

connection.commit()

# Create main summary table
create_summary_table = """
CREATE TABLE GDELT_UA_DAILY_SUMMARY (
    ID VARCHAR(50) PRIMARY KEY,
    EVENT_DATE DATE NOT NULL,
    TOTAL_NO_OF_EVENTS INTEGER,
    NO_OF_EVENTS_UA INTEGER,
    GOLDSTEIN_MEAN DOUBLE,
    GOLDSTEIN_MEDIAN DOUBLE,
    GOLDSTEIN_STD DOUBLE,
    GOLDSTEIN_MIN DOUBLE,
    GOLDSTEIN_MAX DOUBLE,
    AVGTONE_MEAN DOUBLE,
    AVGTONE_MEDIAN DOUBLE,
    AVGTONE_STD DOUBLE,
    AVGTONE_MIN DOUBLE,
    AVGTONE_MAX DOUBLE,
    UPLOAD_TIMESTAMP TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
"""

# Create CAMEO detail table
create_cameo_table = """
CREATE TABLE GDELT_UA_DAILY_CAMEOS (
    ID VARCHAR(100) PRIMARY KEY,
    SUMMARY_ID VARCHAR(50) NOT NULL,
    EVENT_DATE DATE NOT NULL,
    CAMEO_DESCRIPTION VARCHAR(500),
    EVENT_COUNT INTEGER,
    RANK_POSITION INTEGER,
    UPLOAD_TIMESTAMP TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY (SUMMARY_ID) REFERENCES GDELT_UA_DAILY_SUMMARY(ID) ON DELETE CASCADE
)
"""

try:
    cursor.execute(create_summary_table)
    print("✓ Created GDELT_UA_DAILY_SUMMARY table")
except Exception as e:
    print(f"Summary table: {e}")

try:
    cursor.execute(create_cameo_table)
    print("✓ Created GDELT_UA_DAILY_CAMEOS table")
except Exception as e:
    print(f"CAMEO table: {e}")

connection.commit()
cursor.close()
connection.close()

print("\nTables created successfully!")
```

```python
from datetime import date, timedelta
import calendar
def date_slug(d: date) -> str:
    # format date correctly
    return f"{calendar.month_name[d.month].lower()}-{d.day}-{d.year}"

## All above already in py_scheduler

import gdelt

def get_ua_gdelt_data():

    yesterday = date.today() - timedelta(days=1) # Set yesterday as date of interest
    
    gd = gdelt.gdelt()
    events_all = gd.Search([date_slug(yesterday)],table='events',output='gpd',normcols=True, # Replaced 2025 November 24
                        coverage=True) # coverage=True to query all events

    print(events_all.shape)
    events_ua = events_all[(events_all.actiongeocountrycode == 'UP')] # | (events_all.actiongeocountrycode == 'RS')  # Russia
    print(events_ua.shape)

    return events_all, events_ua
```

```python
# Function to insert daily summary data
def insert_daily_summary(events_all, events_ua):
    """
    Insert daily GDELT summary into HANA tables
    
    Args:
        events_all: GeoDataFrame with all events
        events_ua: GeoDataFrame with Ukraine-specific events
    """
    connection = dbapi.connect(address=address, port=port, user=user, password=password)
    cursor = connection.cursor()
    
    # Generate unique ID
    yesterday = date.today() - timedelta(days=1)
    event_date = yesterday.strftime('%Y%m%d')  # YYYYMMDD format
    summary_id = f"GDELT-UA-{event_date}"
    
    # Calculate summary statistics
    total_events = len(events_all)
    ua_events = len(events_ua)
    
    goldstein_stats = events_ua.goldsteinscale.describe()
    avgtone_stats = events_ua.avgtone.describe()
    
    # Insert main summary
    insert_summary = """
    INSERT INTO GDELT_UA_DAILY_SUMMARY (
        ID, EVENT_DATE, TOTAL_NO_OF_EVENTS, NO_OF_EVENTS_UA,
        GOLDSTEIN_MEAN, GOLDSTEIN_MEDIAN, GOLDSTEIN_STD, GOLDSTEIN_MIN, GOLDSTEIN_MAX,
        AVGTONE_MEAN, AVGTONE_MEDIAN, AVGTONE_STD, AVGTONE_MIN, AVGTONE_MAX
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """
    
    cursor.execute(insert_summary, (
        summary_id, event_date, total_events, ua_events,
        float(goldstein_stats['mean']), float(goldstein_stats['50%']), 
        float(goldstein_stats['std']), float(goldstein_stats['min']), float(goldstein_stats['max']),
        float(avgtone_stats['mean']), float(avgtone_stats['50%']), 
        float(avgtone_stats['std']), float(avgtone_stats['min']), float(avgtone_stats['max'])
    ))
    
    # Get top 10 CAMEO codes
    top_10_cameos = events_ua.cameocodedescription.value_counts().head(10)
    
    # Insert CAMEO details
    insert_cameo = """
    INSERT INTO GDELT_UA_DAILY_CAMEOS (
        ID, SUMMARY_ID, EVENT_DATE, CAMEO_DESCRIPTION, EVENT_COUNT, RANK_POSITION
    ) VALUES (?, ?, ?, ?, ?, ?)
    """
    
    for rank, (cameo_desc, count) in enumerate(top_10_cameos.items(), start=1):
        cameo_id = f"{summary_id}_CAMEO_{rank}"
        cursor.execute(insert_cameo, (
            cameo_id, summary_id, event_date, cameo_desc, int(count), rank
        ))
    
    connection.commit()
    cursor.close()
    connection.close()
    
    print(f"✓ Inserted daily summary for {event_date}")
    print(f"  - Total events: {total_events}")
    print(f"  - UA events: {ua_events}")
    print(f"  - Top 10 CAMEOs inserted")
    
    return summary_id
```

```python
# Example usage with current data
events_all, events_ua = get_ua_gdelt_data()
summary_id = insert_daily_summary(events_all, events_ua)
summary_id
```

## Querying from HANA for RAG

```python
def get_latest_gdelt_summary():
    """
    Retrieve the latest GDELT summary and CAMEO data from HANA
    
    Returns:
        tuple: (summary_dict, cameos_dict)
            - summary_dict: Dictionary with latest summary statistics
            - cameos_dict: Dictionary with top 3 CAMEO descriptions as keys and counts as values
    """
    connection = dbapi.connect(address=address, port=port, user=user, password=password)
    cursor = connection.cursor()
    
    # Get latest summary
    summary_query = """
    SELECT 
        ID, EVENT_DATE, TOTAL_NO_OF_EVENTS, NO_OF_EVENTS_UA,
        GOLDSTEIN_MEAN, GOLDSTEIN_MEDIAN, GOLDSTEIN_STD, GOLDSTEIN_MIN, GOLDSTEIN_MAX,
        AVGTONE_MEAN, AVGTONE_MEDIAN, AVGTONE_STD, AVGTONE_MIN, AVGTONE_MAX,
        UPLOAD_TIMESTAMP
    FROM GDELT_UA_DAILY_SUMMARY
    ORDER BY EVENT_DATE DESC
    LIMIT 1
    """
    
    cursor.execute(summary_query)
    summary_row = cursor.fetchone()
    
    if not summary_row:
        cursor.close()
        connection.close()
        return None, None
    
    # Build summary dictionary
    summary_dict = {
        'id': summary_row[0],
        'event_date': summary_row[1],
        'total_events': summary_row[2],
        'ua_events': summary_row[3],
        'goldstein': {
            'mean': summary_row[4],
            'median': summary_row[5],
            'std': summary_row[6],
            'min': summary_row[7],
            'max': summary_row[8]
        },
        'avgtone': {
            'mean': summary_row[9],
            'median': summary_row[10],
            'std': summary_row[11],
            'min': summary_row[12],
            'max': summary_row[13]
        },
        'upload_timestamp': summary_row[14]
    }
    
    # Get corresponding CAMEO codes
    cameo_query = """
    SELECT 
        CAMEO_DESCRIPTION, EVENT_COUNT
    FROM GDELT_UA_DAILY_CAMEOS
    WHERE SUMMARY_ID = ?
    ORDER BY RANK_POSITION ASC
    LIMIT 3
    """
    
    cursor.execute(cameo_query, (summary_dict['id'],))
    cameo_rows = cursor.fetchall()
    
    # Build simple dictionary with top_1, top_2, top_3 as keys
    cameos_dict = {}
    for idx, row in enumerate(cameo_rows, start=1):
        cameos_dict[f'top_{idx}'] = {'description': row[0], 'count': row[1]}
    
    cursor.close()
    connection.close()
    
    return summary_dict, cameos_dict
```

```python
# Test the function
summary, cameos = get_latest_gdelt_summary()
print("Latest Summary:")
print(f"  Date: {summary['event_date']}")
print(f"  Total Events: {summary['total_events']}")
print(f"  UA Events: {summary['ua_events']}")
print(f"  Goldstein Mean: {summary['goldstein']['mean']:.2f}")
print(f"  AvgTone Mean: {summary['avgtone']['mean']:.2f}")
print(f"\nTop 3 CAMEOs:")
for key, value in cameos.items():
    print(f"  {key}: {value['description']} ({value['count']} events)")
```
